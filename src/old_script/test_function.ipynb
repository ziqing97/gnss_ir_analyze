```python
from datetime import datetime,timedelta,timezone
import os

from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

import lib_data_process as dapr
import lib_data_filter as dafi
```

```python
azimut_mask = [270,330]
elevation_mask = [5,30]
min_height = 2
max_height = 4
t_range = 30
trigger_list = ['snr1','snr2','snr5','snr7','snr8']
```

```python
main_path_2 = os.path.abspath('../data/20220811/2/')
main_path_3 = os.path.abspath('../data/20220811/3/')

starttime = datetime(year=2022,month=8,day=11,hour=8,\
        minute=0,second=0,tzinfo=timezone.utc)
endtime = datetime(year=2022,month=8,day=11,hour=14,\
    minute=0,second=0,tzinfo=timezone.utc)

ts_0811_2 = dapr.generate_timeseries(main_path_2, azimut_mask, elevation_mask, min_height, max_height, t_range, starttime, endtime, trigger_list, 1)
ts_0811_3 = dapr.generate_timeseries(main_path_3, azimut_mask, elevation_mask, min_height, max_height, t_range, starttime, endtime, trigger_list, 1)
```

```python
fig,ax = plt.subplots()
ax.scatter(ts_0811_2.keys(),ts_0811_2.values())
ax.scatter(ts_0811_3.keys(),ts_0811_3.values())
```
