```python
import numpy as np
from scipy import signal
from scipy.fftpack import fft,fftfreq
import matplotlib.pyplot as plt
```

```python
a = np.arange(0,100,0.1)
na = np.random.normal(0,0.1,1000)
aa = np.sin(a)+na

b = np.arange(0,100,0.2)
nb = np.random.normal(0,0.1,500)
bb = np.sin(b)+nb

c = np.arange(0,100,0.4)
nc = np.random.normal(0,0.1,250)
cc = np.sin(c)+nc
```

```python
plt.plot(a,np.sin(a))
plt.plot(b,np.sin(b))
plt.plot(c,np.sin(c))
```

```python
plt.plot(a,aa)
plt.plot(b,bb)
plt.plot(c,cc)
```

```python
freq = np.arange(0.1,1.5,0.001)
powera = signal.lombscargle(a,aa,freq)
powerb = signal.lombscargle(b,bb,freq)
powerc = signal.lombscargle(c,cc,freq)

```

```python
fig,ax = plt.subplots()
ax.plot(freq/2/np.pi,powera)
ax.plot(freq/2/np.pi,powerb)
ax.plot(freq/2/np.pi,powerc)
```

```python
poweraa = fft(aa)
powerbb = fft(bb)
powercc = fft(cc)

fa = fftfreq(len(aa),1/10)
fb = fftfreq(len(bb),1/5)
fc = fftfreq(len(cc),1/2.5)
```

```python
fig,ax = plt.subplots()
ax.plot(fa,poweraa)
ax.plot(fb,powerbb)
ax.plot(fc,powercc)
ax.set_xlim([0,0.3])
```

```python
left = 0.995
right = 1.005
```

```python

area_index = np.bitwise_and(freq>=left,freq<=right)
area_x = np.diff(freq[area_index])
area_y_origin = (powera[area_index][0:-1]+powera[area_index][1:])/2
scale = np.sum(np.multiply(area_x,area_y_origin))
area_y = area_y_origin/scale
i = 0
while np.sum(np.multiply(area_y[i:len(area_y)-i],area_x[i:len(area_y)-i]))>0.68:
    i+=1
    if i>len(area_y)/2-1:
        break
uca = np.abs(1-freq[area_index][i])
uca
```
