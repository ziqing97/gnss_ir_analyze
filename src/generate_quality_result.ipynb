```python
import os
from datetime import datetime,timezone,timedelta

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from tqdm import tqdm
from scipy import signal

import lib_data_filter as dafi
import lib_data_process as dapr
import lib_plot

import pickle
```

```python
min_height = 2
max_height = 4
azimut_mask = [280,320]
```

```python
main_path_2 = os.path.abspath('../data/20220811/2/')
main_path_3 = os.path.abspath('../data/20220811/3/')

ts_dict = {}
max_ele_settings = [20,25,30,40,50]
t_range_settings = [15,20,25,30,60]
for ele in max_ele_settings:
    ts_dict[ele] = {}
    for t_range in t_range_settings:
        ts_dict[ele][t_range] = {}
        elevation_mask = [3,ele]
        ts_dict[ele][t_range][2] = dapr.generate_timeseries(main_path_2, azimut_mask, elevation_mask, min_height, max_height, t_range,8,11)
        ts_dict[ele][t_range][3] = dapr.generate_timeseries(main_path_3, azimut_mask, elevation_mask, min_height, max_height, t_range,8,11)

with open('20220811.pkl', 'wb') as f:
    pickle.dump(ts_dict, f)
```
